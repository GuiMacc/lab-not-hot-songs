```python
import numpy as np
import pandas as pd
```

```python
big_music = pd.read_csv("tcc_ceds_music.csv")
big_music.head()
```

```python
smaller_music = big_music.sample(4000, random_state=1).reset_index(drop=True)
smaller_music.head()
```

```python
print(smaller_music.shape)
print(big_music.shape)
```

```python
display(smaller_music['genre'].value_counts()/len(smaller_music))
display(big_music['genre'].value_counts()/len(big_music))
```

```python
not_hot100=smaller_music[['track_name', 'artist_name']].rename(columns = ({"track_name": "song_name", "artist_name":"song_artist"}))
print(not_hot100.shape)
display(not_hot100.head())
```

```python
hot100=pd.read_csv("hot100.csv")
```

```python
hot100.head()
```

```python
hot100 = hot100.applymap(lambda s:s.lower() if type(s) == str else s)
hot100.head()
```

```python
repeated = hot100.merge(not_hot100.reset_index())['index'].tolist()
repeated
```
